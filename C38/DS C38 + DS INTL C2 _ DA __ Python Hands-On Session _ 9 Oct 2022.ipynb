```python
# Madhura

string ='You love python'
answer = []
vowels = set(['a','e','i','o','u'])

for i in string:
    if i.lower() in vowels:
        answer.append(i)

for i in string:
    if i.lower() not in vowels:
        answer.append(i)
print(''.join(answer))
```

```python
import sys

class Solution:
    st = []
    qu = []
    
    def pushCharacter(self, s):
        self.st.append(s)
        
    def enqueueCharacter(self, s):
        self.qu.append(s)
        
    def popCharacter(self):
        return self.st.pop()
    
    def dequeueCharacter(self):
        ch = self.qu[0]
        del self.qu[0]
        return ch
    
    
s = input() # NITIN
obj = Solution()

l = len(s) # len = 5

# push / enqueue all the characters of string s to stack / queue
for i in range(l):
    obj.pushCharacter(s[i])
    obj.enqueueCharacter(s[i])
    
isPalindrome = True

'''
pop the top character from the stack
dequeue the first character from the queue
compare both the characters
'''

for i in range(l // 2):
    if obj.popCharacter() != obj.dequeueCharacter():
        isPalindrome = False
        break
        
if isPalindrome == True:
    print(s,"is a Palindrome")
else:
    print(s,"is not a Palindrome")
```
